```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
binomialvar = np.random.binomial(10, .5, [100])
chisquarevar = np.random.chisquare(3,[100])
gammavar = np.random.gamma(3, 1, [100])
noncentral_chisquarevar = np.random.noncentral_chisquare(3, 1, [100])
poissonvar = np.random.poisson(6,[100])
exponentialvar = np.random.exponential(1,[100])
```

```python
variables = pd.DataFrame()
variables['binomial'] = binomialvar
variables['chisquare'] = chisquarevar
variables['gamma'] = gammavar
variables['noncentral'] = noncentral_chisquarevar
variables['poisson'] = poissonvar
variables['exponential'] = exponentialvar
varnames = ('binomial', 'chisquare', "gamma", 'noncentral', 'poisson', 'exponential')
```

```python
for variable in varnames:
    plt.hist(variables[variable])
    plt.axvline(variables[variable].mean(), color='b', linestyle='solid', linewidth=2)
    plt.axvline(variables[variable].mean() + variables[variable].std(), color='b', linestyle='dashed', linewidth=2)
    plt.axvline(variables[variable].mean() - variables[variable].std(), color='b', linestyle='dashed', linewidth=2)
    plt.show()
    

```

The binomial, poisson and gamma distributions feature some clustering around the mean, although each is skewed slightly differently.  This makes them better described by the mean and standard deviation.  The exponential, chi square and noncentral chi square all feature large amounts of data more than one standard deviation above the mean, making these statistics less descriptive.  These three functions appear to more closely resemble exponential decay than a normal distribution.

```python
rand1 = np.random.normal(5, .5, 1000)
rand2 = np.random.normal(10, 1, 1000)
randsum = rand1 + rand2
plt.hist(randsum)
plt.axvline(randsum.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(randsum.mean() + randsum.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(randsum.mean() - randsum.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

It generally resembles a normal distribution with the mean being the sum of the means of the first two data sets and the standard deviation being closer to the square root of the sum of the standard deviations of the first two sets.  Its slightly negatively skewed.
